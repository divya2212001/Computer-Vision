```python
from ultralytics import YOLO
import glob
import os

model = YOLO("yolov8s.pt")

image_paths = sorted(glob.glob("frames/*.png"))

os.makedirs("output_images", exist_ok=True)

for img in image_paths:
    results = model(img)
    results[0].save(filename=os.path.join("output_images", os.path.basename(img)))

print("Batch detection done!")
```

```python
!ffmpeg -framerate 30 -i output_images/frame_%04d.png -c:v libx264 -pix_fmt yuv420p detected_video.mp4
```
